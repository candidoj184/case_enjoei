# Tratamentos da base

Remoção dos valores guardados em dicionários na coluna "rating" e os transformando em colunas

```python
# extração  productId e quantity de cada dicionário

'''São Criadas  novas colunas chamadas 'count' e 'rate' O apply() executa a função lambda para cada linha.
A função lambda lambda x: x['count'] pega cada dicionário (x) e extrai o valor da chave 'count' gerando uma nova coluna.
O mesmo se repete para 'rate' '''

df_products['count'] = df_products['rating'].apply(lambda x: x['count'])
df_products['rate'] = df_products['rating'].apply(lambda x: x['rate'])
```

```python
# Remoção da coluna rating original
df_products = df_products.drop('rating', axis=1)
```
