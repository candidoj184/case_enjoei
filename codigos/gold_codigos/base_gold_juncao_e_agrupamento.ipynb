## Imports

```python
import pandas as pd
```

## Leitura das bases da camada silver

```python
df_users = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/GitHub/case_enjoei/datasets/silver/users.parquet')
df_produts = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/GitHub/case_enjoei/datasets/silver/products.parquet')
df_carts = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/GitHub/case_enjoei/datasets/silver/carts.parquet')
```

## Junções das bases

### Join da tabela de usuários com a tabela de carrinho para trazer informações de carrinho dos usuários

```python
df_merged = df_carts.merge(df_users, left_on='user_id', right_on='user_id', how='left')
```

```python
df_merged
```

### Join da tabela de usuários + carrinho com a tabela de produts para trazer informações dos produtos

```python
df_final = df_merged.merge(df_produts,left_on='product_id', right_on='product_id', how='left')
```

### Conferência dos dados

```python
df_final
```

### Ordenação das colunas do dataframe unificado para melhor entendimento

```python
colunas_ordenadas = [
    'user_id',
    'full_name',
    'email',
    'username',
    'password',
    'phone',
    'city',
    'zip_code',
    'street',
    'number',
    'geolocation_lat',
    'geolocation_long',
    'product_id',
    'category',
    'quantity',
    'price',
    'date',
    'count',
    'rate',
    'title',
    'description',
    'image',
    'carts_id'
]

df_final = df_final[colunas_ordenadas]
```
