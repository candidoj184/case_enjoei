## Imports

```python
import pandas as pd
```

## Leitura das bases da camada silver

```python
df_base_users = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/Enjoei/Case_enjoei/datasets/silver/users.parquet')
df_base_produts = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/Enjoei/Case_enjoei/datasets/silver/products.parquet')
df_base_carts = pd.read_parquet('C:/Users/candi/OneDrive/Documentos/Enjoei/Case_enjoei/datasets/silver/carts.parquet')
```

## Junções das bases

### Join da tabela de usuários com a tabela de carrinho para trazer informações de carrinho dos usuários

```python
df_merged = df_base_carts.merge(df_base_users, left_on='userId', right_on='id', how='left')
```

### Join da tabela de usuários + carrinho com a tabela de produts para trazer informações dos produtos

```python
df_final = df_merged.merge(df_base_produts,left_on='productId', right_on='id', how='left')
```

### Remoção de resquícios do join

```python
df_final = df_final.drop('id_x', axis=1)
df_final = df_final.drop('id_y', axis=1)
df_final = df_final.drop('id', axis=1)
```

#### Ordenação das colunas do dataframe unificado para melhor entendimento

```python
colunas_ordenadas = [
    'userId',
    'fullname',
    'email',
    'username',
    'password',
    'phone',
    'city',
    'zipcode',
    'street',
    'number',
    'geolocation_lat',
    'geolocation_long',
    'productId',
    'category',
    'quantity',
    'price',
    'date',
    'count',
    'rate',
    'title',
    'description',
    'image'
]

df_final = df_final[colunas_ordenadas]
```

### Conferência da ordenação

```python
df_final
```

## Agrupamentos

```python
df_analysis = df_final.groupby('userId').agg({
    'date': 'max',
    'category': lambda x: x.value_counts().index[0], 'fullname': 'first'
})
```

```python
# 2. Calcular quantidade de produtos por categoria para cada usuário
category_counts = df_final.groupby(['userId', 'category']).size().reset_index(name='count')
top_categories = category_counts.sort_values('count', ascending=False).groupby('userId').first()
```

```python
result = df_analysis.merge(
   top_categories[['category', 'count']], 
   left_index=True, 
   right_index=True,
   how='left',
   suffixes=('', '_removida')
)

```

```python
result = result.drop('category_removida', axis=1)
```

```python
result = result.reset_index()
result = result[['userId', 'fullname', 'date', 'category', 'count']]
result
```
